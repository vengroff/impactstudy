# Cubic Target; Variable Number of Cubic Terms

Uniformly distributed $x_i$.

$$
\begin{align}
f(X) & = A \cdot X[0, m_1) + b + \left(\sum_{i = m_1}^{m_1 + m_2} {\beta_i} + {\alpha_i}_{\chi[t_i, \infty)}(X[i]) \right); \\ 
m_2 & = 4; \\
m_1 & \ge 1; \\
s & = 5; \\
Z & = N(0, \sigma^2).
\end{align}
$$
